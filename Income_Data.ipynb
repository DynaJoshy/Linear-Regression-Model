```python
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("canada_income.csv")
```

```python
df
```

```python

plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.title('Per Capita Income Over Years')  # Optional: Adding a title

# Creating a scatter plot
plt.scatter(df['year'], df['per capita income (US$)'], color='red', marker='+')

# Display the plot
plt.show()
```

linear regresson : fit accepts first parameter as to data upon to predict and the second to be predicted

```python
df = df.dropna()


reg = linear_model.LinearRegression()

reg.fit(df[['year']], df['per capita income (US$)'])
```

```python
reg.predict([[2150]])
```

```python
pre_year=pd.read_csv("New.csv")
```

```python
p=reg.predict(pre_year)
p
#alternatively,when csv file has one filled values and another blank values
#p=red.predict(pre_year['per capita'])
```
